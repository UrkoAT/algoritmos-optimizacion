```python
#N Reinas - Vuelta AtrÃ¡s()
################################################################################

#Verifica que en la soluciÃ³n parcial no hay amenzas entre reinas
################################################################
def es_prometedora(SOLUCION,etapa):
################################################################
  #print(SOLUCION)
  #Si la soluciÃ³n tiene dos valores iguales no es valida => Dos reinas en la misma fila
  for i in range(etapa+1):
    #print("El valor " + str(SOLUCION[i]) + " estÃ¡ " +  str(SOLUCION.count(SOLUCION[i])) + " veces")
    if SOLUCION.count(SOLUCION[i]) > 1:
      return False

    #Verifica las diagonales
    for j in range(i+1, etapa +1 ):
      #print("Comprobando diagonal de " + str(i) + " y " + str(j))
      if abs(i-j) == abs(SOLUCION[i]-SOLUCION[j]) : return False
  return True

#Traduce la soluciÃ³n al tablero
################################################################
def escribe_solucion(S):
################################################################
  n = len(S)
  for x in range(n):
    print("")
    for i in range(n):
      if S[i] == x+1:
        print(" X " , end="")
      else:
        print(" - ", end="")



#Proceso principal de N-Reinas
################################################################
def reinas(N, solucion=[],etapa=0):
################################################################
### ....
  if len(solucion) == 0:         # [0,0,0...]
    solucion = [0 for i in range(N) ]

  for i in range(1, N+1):
    solucion[etapa] = i
    if es_prometedora(solucion, etapa):
      if etapa == N-1:
        print(solucion)
      else:
        reinas(N, solucion, etapa+1)
    else:
      None

  solucion[etapa] = 0

reinas(8,solucion=[],etapa=0)
```

```python
#Viaje por el rio - ProgramaciÃ³n dinÃ¡mica
################################################################################

TARIFAS = [
[0,5,4,3,999,999,999],
[999,0,999,2,3,999,11],
[999,999, 0,1,999,4,10],
[999,999,999, 0,5,6,9],
[999,999, 999,999,0,999,4],
[999,999, 999,999,999,0,3],
[999,999,999,999,999,999,0]
]

#999 se puede sustituir por float("inf")


#Calculo de la matriz de PRECIOS y RUTAS
################################################################
def Precios(TARIFAS):
################################################################
  #Total de Nodos
  N = len(TARIFAS[0])

  #InicializaciÃ³n de la tabla de precios
  PRECIOS = [ [9999]*N for i in [9999]*N]
  RUTA = [ [""]*N for i in [""]*N]

  for i in range(0,N-1):
    RUTA[i][i] = i             #Para ir de i a i se "pasa por i"
    PRECIOS[i][i] = 0          #Para ir de i a i se se paga 0
    for j in range(i+1, N):
      MIN = TARIFAS[i][j]
      RUTA[i][j] = i

      for k in range(i, j):
        if PRECIOS[i][k] + TARIFAS[k][j] < MIN:
            MIN = min(MIN, PRECIOS[i][k] + TARIFAS[k][j] )
            RUTA[i][j] = k          #Anota que para ir de i a j hay que pasar por k
        PRECIOS[i][j] = MIN

  return PRECIOS,RUTA
################################################################

PRECIOS,RUTA = Precios(TARIFAS)
#print(PRECIOS[0][6])

print("PRECIOS")
for i in range(len(TARIFAS)):
  print(PRECIOS[i])

print("\nRUTA")
for i in range(len(TARIFAS)):
  print(RUTA[i])

#Determinar la ruta con Recursividad
def calcular_ruta(RUTA, desde, hasta):
  if desde == hasta:
    #print("Ir a :" + str(desde))
    return ""
  else:
    return str(calcular_ruta( RUTA, desde, RUTA[desde][hasta])) +  \
                ',' + \
                str(RUTA[desde][hasta] \
              )

print("\nLa ruta es:")
calcular_ruta(RUTA, 0,6)
```

Bucle i: n - 1 veces
Bucle j: n - i - 1 veces

Entonces la complejidad seria la suma de n - 1, n - 2, n -3... -> Lo que equivale a la funcion geometrica (n*(n-1)) / 2 -> O(n^2)

Aplicando la estrategia divide y vencerÃ¡s, podremos primero ordenar la array, lo cual usando algoritmos como quicksort o mergesort nos darian un coste de O(n log n) y luego iterar solo una vez la array O(n) en busca de la diferencia minima, ya que estaran uno al lado del otro. Por lo que quedariamos con un coste de O(n log n)

```python
def encontrar_puntos_mas_cercanos_divide_venceras(arr):
    arr.sort()

    min_diff = float('inf')
    puntos_mas_cercanos = []

    for i in range(len(arr) - 1):
        diff = arr[i+1] - arr[i]

        if diff < min_diff:
            min_diff = diff
            puntos_mas_cercanos = (arr[i], arr[i+1])

    return puntos_mas_cercanos
```

Creo que no se podria mejorar mucho mas. AdemÃ¡s para hacer la version 2D y 3D, no podremos ordenar linealmente los puntos y se necesitaran implementaciones mas avanzadas:

```python
import math

def euclidean_distance(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2)

def closest_pair_in_strip(strip, d, best_pair):
    min_d = d
    for i in range(len(strip)):
        j = i + 1
        while j < len(strip) and (strip[j][1] - strip[i][1]) < min_d:
            dist = euclidean_distance(strip[i], strip[j])
            if dist < min_d:
                min_d = dist
                best_pair = (strip[i], strip[j])
            j += 1

    return min_d, best_pair

def closest_pair_recursive(points_sorted_x, points_sorted_y):
    if len(points_sorted_x) <= 3:
        min_d = float('inf')
        best_pair = None
        for i in range(len(points_sorted_x)):
            for j in range(i+1, len(points_sorted_x)):
                dist = euclidean_distance(points_sorted_x[i], points_sorted_x[j])
                if dist < min_d:
                    min_d = dist
                    best_pair = (points_sorted_x[i], points_sorted_x[j])
        return min_d, best_pair

    mid = len(points_sorted_x) // 2
    mid_x = points_sorted_x[mid][0]

    points_left = points_sorted_x[:mid]
    points_right = points_sorted_x[mid:]

    points_left_y = list(filter(lambda point: point[0] <= mid_x, points_sorted_y))
    points_right_y = list(filter(lambda point: point[0] > mid_x, points_sorted_y))

    d_left, pair_left = closest_pair_recursive(points_left, points_left_y)
    d_right, pair_right = closest_pair_recursive(points_right, points_right_y)

    if d_left < d_right:
        d = d_left
        best_pair = pair_left
    else:
        d = d_right
        best_pair = pair_right

    strip = [p for p in points_sorted_y if abs(p[0] - mid_x) < d]

    d_strip, pair_strip = closest_pair_in_strip(strip, d, best_pair)
    if d_strip < d:
        return d_strip, pair_strip

    return d, best_pair

def closest_pair(points):
    points_sorted_x = sorted(points, key=lambda p: p[0])
    points_sorted_y = sorted(points, key=lambda p: p[1])

    _, best_pair = closest_pair_recursive(points_sorted_x, points_sorted_y)
    return best_pair
```

```python
import math

def euclidean_distance_3d(p1, p2):
    return math.sqrt((p1[0] - p2[0])**2 + (p1[1] - p2[1])**2 + (p1[2] - p2[2])**2)

def closest_pair_in_strip_3d(strip, d, best_pair):
    min_d = d
    # Sort the strip by y-coordinate for efficient comparison
    strip.sort(key=lambda p: p[1])
    for i in range(len(strip)):
        # Check only points within d in the y and z directions
        for j in range(i+1, len(strip)):
            if strip[j][1] - strip[i][1] < min_d:
                dist = euclidean_distance_3d(strip[i], strip[j])
                if dist < min_d:
                    min_d = dist
                    best_pair = (strip[i], strip[j])
            else:
                break
    return min_d, best_pair

def closest_pair_recursive_3d(points_sorted_x, points_sorted_y, points_sorted_z):
    if len(points_sorted_x) <= 3:
        min_d = float('inf')
        best_pair = None
        for i in range(len(points_sorted_x)):
            for j in range(i+1, len(points_sorted_x)):
                dist = euclidean_distance_3d(points_sorted_x[i], points_sorted_x[j])
                if dist < min_d:
                    min_d = dist
                    best_pair = (points_sorted_x[i], points_sorted_x[j])
        return min_d, best_pair

    mid = len(points_sorted_x) // 2
    mid_point = points_sorted_x[mid]

    points_left_x = points_sorted_x[:mid]
    points_right_x = points_sorted_x[mid:]
    points_left_y = list(filter(lambda p: p[0] <= mid_point[0], points_sorted_y))
    points_right_y = list(filter(lambda p: p[0] > mid_point[0], points_sorted_y))
    points_left_z = list(filter(lambda p: p[0] <= mid_point[0], points_sorted_z))
    points_right_z = list(filter(lambda p: p[0] > mid_point[0], points_sorted_z))

    d_left, pair_left = closest_pair_recursive_3d(points_left_x, points_left_y, points_left_z)
    d_right, pair_right = closest_pair_recursive_3d(points_right_x, points_right_y, points_right_z)

    d = min(d_left, d_right)
    best_pair = pair_left if d_left < d_right else pair_right

    # Create a strip in 3D, considering only points within d of the dividing plane
    strip = [p for p in points_sorted_y if abs(p[0] - mid_point[0]) < d]

    d_strip, pair_strip = closest_pair_in_strip_3d(strip, d, best_pair)
    if d_strip < d:
        return d_strip, pair_strip

    return d, best_pair

def closest_pair_3d(points):
    points_sorted_x = sorted(points, key=lambda p: p[0])
    points_sorted_y = sorted(points, key=lambda p: p[1])
    points_sorted_z = sorted(points, key=lambda p: p[2])

    _, best_pair = closest_pair_recursive_3d(points_sorted_x, points_sorted_y, points_sorted_z)
    return best_pair
```
